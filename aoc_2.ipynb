```python
def read_file(file_path):
    
    reports = []
    
    with open(file_path, 'r') as file:
        for line in file:
            levels = [int(x) for x in line.split()]
            reports.append(levels)

    
    return reports

file_path = 'aoc2.txt'
reports = read_file(file_path)
```

```python
def check_diff(a : int, b : int) -> bool:

    diff = abs(a - b)

    return 1 <= diff <= 3


def check_report(report : list) -> bool:

    # Condition 1

    increasing  = all([report[i] < report[i+1] for i in range(0, len(report)-1)])

    decreasing  = all([report[i] > report[i+1] for i in range(0, len(report)-1)])

    condition_1 = increasing or decreasing


    # Condition 2

    condition_2 = all([check_diff(report[i],report[i+1]) for i in range(0, len(report)-1)])

    return condition_1 and condition_2
```

```python
safe_reports = sum([check_report(r) for r in reports])

safe_reports
```
